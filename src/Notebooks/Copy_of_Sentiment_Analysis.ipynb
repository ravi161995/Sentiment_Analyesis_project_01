```python
import os
import pandas as pd
import tensorflow as tf
import numpy as np
from transformers import DistilBertTokenizerFast, TFDistilBertForSequenceClassification
from sklearn.metrics import classification_report
```

```python
# Define a directory to save the model
save_directory = "./sentiment_model"

# Initialize and compile the model if not already done
# Check if 'model' is defined and is an instance of the expected class
if 'model' not in locals() or not isinstance(model, TFDistilBertForSequenceClassification):
    print("Initializing and compiling the model...")
    MODEL_NAME = "distilbert-base-uncased"
    model = TFDistilBertForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2, from_pt=True)
    optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    print("Model initialized and compiled.")
else:
    print("Model already defined and compiled.")


# Save the model
model.save_pretrained(save_directory)

print(f"Model saved to {save_directory}")
```

To fix the "Invalid Notebook" error in Google Colab, we can access and modify the notebook's metadata directly. We'll use the `nbformat` library to read the notebook file, remove the problematic 'widgets' key from the metadata, and then save the modified notebook.

```python
import nbformat
import os
import ipynbname

# Get the path to the current notebook file dynamically
try:
    # In Colab, the notebook path can be accessed via the environment variable
    # This is a more reliable way to get the path in this environment.
    notebook_path = os.environ.get('NOTEBOOK_NAME')
    if notebook_path:
        # Colab's NOTEBOOK_NAME environment variable only provides the filename,
        # so we need to construct the full path if necessary.
        # However, for modifying the current notebook in Colab, using the filename
        # with nbformat's `read` and `write` functions is often sufficient
        # as they operate on the current notebook context.
        # If the file wasn't found, this means nbformat couldn't handle the name directly.
        # We'll rely on the direct name for now and handle the error if it occurs.
        print(f"Attempting to use notebook filename from environment: {notebook_path}")
    else:
        print("Error: Could not determine notebook name from environment.")
        notebook_path = None

except Exception as e:
    print(f"An error occurred while trying to determine notebook path: {e}")
    notebook_path = None


if notebook_path:
    try:
        # Read the notebook
        with open(notebook_path, 'r', encoding='utf-8') as f:
            notebook = nbformat.read(f, as_version=4)

        # Check if 'widgets' exists in metadata and remove it
        if 'widgets' in notebook['metadata']:
            del notebook['metadata']['widgets']
            print("Removed 'widgets' from notebook metadata.")
        else:
            print("'widgets' not found in notebook metadata. No changes made.")

        # Write the modified notebook back to the file
        with open(notebook_path, 'w', encoding='utf-8') as f:
            nbformat.write(notebook, f)

        print("Notebook metadata updated. You can now try uploading to GitHub.")

    except FileNotFoundError:
        print(f"Error: Notebook file not found at {notebook_path}. This might happen if the environment variable doesn't provide a directly accessible path. Manual download and upload might be needed.")
    except Exception as e:
        print(f"An error occurred: {e}")
else:
    print("Could not proceed with metadata update as notebook path was not determined.")
```

```python
import pandas as pd

train_data = train[['cleaned_text', 'target']]
test_data = test[['cleaned_text', 'target']]

train_data.head()
```

```python
X_train, y_train = train_data['cleaned_text'].tolist()[:int(len(train_data) * 0.05)], train_data['target'].tolist()[:int(len(train_data) * 0.05)]
X_test, y_test   = test_data['cleaned_text'].tolist()[:int(len(test_data) * 0.05)], test_data['target'].tolist()[:int(len(test_data) * 0.05)]

print(f"Size of X_train: {len(X_train)}")
print(f"Size of y_train: {len(y_train)}")
print(f"Size of X_test: {len(X_test)}")
print(f"Size of y_test: {len(y_test)}")
```

```python
MODEL_NAME = "distilbert-base-uncased"
tokenizer = DistilBertTokenizerFast.from_pretrained(MODEL_NAME)
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True, max_length=128)
test_encodings  = tokenizer(X_test, truncation=True, padding=True, max_length=128)
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((dict(train_encodings), y_train)).shuffle(10000).batch(8)
test_dataset  = tf.data.Dataset.from_tensor_slices((dict(test_encodings), y_test)).batch(8)
```

```python
model = TFDistilBertForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2, from_pt=True)
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
```

```python
model.fit(train_dataset, validation_data=test_dataset, epochs=5)
```

```python
y_pred_logits = model.predict(test_dataset).logits
y_pred = np.argmax(y_pred_logits, axis=1)
print(classification_report(y_test, y_pred, target_names=["Negative", "Positive"]))
```
